## Dummy variable using scikit-learn library

- label encode
- OneHot Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
price_data = data.copy()
price_data.town = le.fit_transform(data.town)
price_data
```

## OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
column_transformer =ColumnTransformer([("encoder", OneHotEncoder(), [0])]
                                      , remainder = 'passthrough')
```

```python
data = column_transformer.fit_transform(data)
data
```

```python
x =pd.DataFrame(data)
x
```

## Splitting dataset

```python
price = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv")
price
```

```python
X = price.iloc[:, 0:2]
X
```

```python
y = price.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```
